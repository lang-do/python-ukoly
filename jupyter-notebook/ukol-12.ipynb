```python
import pandas
import matplotlib.pyplot as plt

file = "platy_2021_02.csv"
plat = pandas.read_csv(file, index_col='cislo_zamestnance')
plat.hist(
            bins = 30,
            range = (30000,60000),
            rwidth = 0.9,
            facecolor = "#000000"
            )
plt.title("Mzdy zaměstnanců\n",
          fontweight ="bold")
plt.xlabel("Výplata [Kč]")
plt.ylabel("Četnost [-]")
plt.show()
```

```python
import requests

url = "https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/zakladni-dotazy/staty.json"
r = requests.get(url) # <Response [200]>
open("staty.json", "wb").write(r.content)
```

```python
import pandas as pd
staty = pd.read_json("staty.json")
staty
```

```python
staty = staty.set_index("name") #index řádku je název státu
```

```python
staty.loc[["Czech Republic", "Yemen"], ["population", "area"]]
```

```python
populace = staty["population"]
populace.sum()
#staty["population"].sum() - v jednom kroku
```

```python
print(f"Prumer populace je {populace.mean()}.")
print(f"Minimum populace je {populace.min()}.")
print(f"Maximum populace je {populace.max()}.")
print(f"Median populace je {populace.median()}.")
```

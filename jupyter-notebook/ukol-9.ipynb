```python
import pandas as pd
data = pd.read_csv("temperature.csv")
data
```

```python
data = data.set_index("City")
data.index
```

```python
data.loc["Prague"]
```

```python
hotdata = data[data["AvgTemperature"] > 80]
hotdata
```

```python
hotEu = data[(data["AvgTemperature"] > 60) & (data["Region"] == "Europe")]
hotEu
```

```python
extremeT = data[(data["AvgTemperature"] > 80) | (data["AvgTemperature"] < -20)]
extremeT
```

```python
import pytemperature
data["AvgTemperatureCelsia"] = pytemperature.f2c(data["AvgTemperature"])
```

```python
hotdataC = data[data["AvgTemperatureCelsia"] > 30]
hotdataC
```

```python
notColdEuC = data[(data["AvgTemperatureCelsia"] > 15) & (data["Region"] == "Europe")]
notColdEuC
```

```python
extremeTinC = data[(data["AvgTemperatureCelsia"] > 30) | (data["AvgTemperatureCelsia"] < -10)]
extremeTinC
```
